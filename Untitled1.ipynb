```python
sym = 'BCH-USD'
BCH_USD_data = pd.DataFrame(data[sym]).dropna()
BCH_USD_data.tail()
```

```python
window = 10
BCH_USD_data['min'] = data[sym].rolling(window=window).min()
BCH_USD_data['mean'] = data[sym].rolling(window=window).mean()
BCH_USD_data['std'] = data[sym].rolling(window=window).std()
BCH_USD_data['median'] = data[sym].rolling(window=window).median()
BCH_USD_data['max'] = data[sym].rolling(window=window).max()['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean()
BCH_USD_data.dropna().head()
```

```python
ax = BCH_USD_data[['min', 'mean', 'max']].iloc[-200:].plot(figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8)
data[sym].iloc[-200:].plot(ax=ax, lw=2.0);
```

```python
BCH_USD_data['SMA1'] = data[sym].rolling(window=20).mean()
BCH_USD_data['SMA2'] = data[sym].rolling(window=60).mean()
BCH_USD_data[[sym, 'SMA1', 'SMA2']].tail()
```

```python
BCH_USD_data[[sym, 'SMA1', 'SMA2']].plot(figsize=(10, 6))
```

```python
BCH_USD_data.dropna(inplace=True)
BCH_USD_data['positions'] = np.where(BCH_USD_data['SMA1'] > BCH_USD_data['SMA2'],1,-1)
ax = BCH_USD_data[[sym, 'SMA1', 'SMA2', 'positions']].plot(figsize=(10, 6),secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```

```python
sprets = np.log(Corr_data / Corr_data.shift(1))
sprets.head()
```

```python
sprets.dropna(inplace=True)
sprets.plot(subplots=True, figsize=(10, 6))
```

```python
pd.plotting.scatter_matrix(sprets, alpha=0.2, diagonal='hist', hist_kwds={'bins': 35}, figsize=(10, 6))
```

```python
reg = np.polyfit(sprets['BCH-USD'], sprets['^VIX'], deg=1)
ax = sprets.plot(kind='scatter', x='BCH-USD', y='^VIX', figsize=(10, 6))
ax.plot(sprets['BCH-USD'], np.polyval(reg, sprets['BCH-USD']), 'r', lw=2);
```
